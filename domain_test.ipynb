# Access Token:

```python
auth_url = 'https://auth.domain.com.au/v1/connect/token'
```

# Agents & Listings：

PS: 找到 登载在、过 DOMAIN 的广告 信息, 提供 agent 信息

PS: 注意，需要 Ad ID !!!!!

```python
# api_listings_read, or api_listings_write

listings_scopes = ['api_listings_read']

 # POST request for token
listings_scopes_response = requests.post(auth_url, data = {
                        'client_id':client_id,
                        'client_secret':client_secret,
                        'grant_type':'client_credentials',
                        'scope':listings_scopes,
                        'Content-Type':'text/json'
                        })
listings_response_token = listings_scopes_response.json() 
listings_access_token = listings_response_token["access_token"]
listings_access_token


property_id = "2013390813" # from Properties & Locations -- properties{id} --- 中获取 图片信息，上面有 Ad ID !!!!!
listings_url = "https://api.domain.com.au/v1/listings/"+property_id


# GET Request 
listings_auth = {"Authorization":"Bearer " + listings_access_token}
listings_auth_request = requests.get(listings_url,headers=listings_auth)
listings_respond = listings_auth_request.json()
listings_respond
```
